```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.patches as patches

from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
    
seaborn_colors = pd.read_csv("Colors.csv")
seaborn_colors = seaborn_colors.columns
seaborn_colors =  list(filter(lambda a: a != 'jet', seaborn_colors))
seaborn_colors =  list(filter(lambda a: a != 'jet_r', seaborn_colors))

for i in range(0,len(seaborn_colors)):
    pal = sns.color_palette(seaborn_colors[i], n_colors=8).as_hex()
    
    fig3 = plt.figure(figsize=(8,1))
    ax3 = fig3.add_subplot(111)
    for p in [
        patches.Rectangle((0, 0), 1/8, 1/8,facecolor= pal[0]),
        patches.Rectangle((1/8, 0), 1/8, 1/8,facecolor= pal[1]),
        patches.Rectangle((2/8, 0), 1/8, 1/8,facecolor= pal[2]),
        patches.Rectangle((3/8, 0), 1/8, 1/8,facecolor= pal[3]),
        patches.Rectangle((4/8, 0), 1/8, 1/8,facecolor= pal[4]),
        patches.Rectangle((5/8, 0), 1/8, 1/8,facecolor= pal[5]),
        patches.Rectangle((6/8, 0), 1/8, 1/8,facecolor= pal[6]),
        patches.Rectangle((7/8, 0), 1/8, 1/8,facecolor= pal[7]),

    ]:
        ax3.set_ylim(0,1/8)
        ax3.add_patch(p)
        ax3.axis('off')
        ax3.set_title(seaborn_colors[i], fontsize=16, loc='left')
```
